```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.preprocessing import OrdinalEncoder

# Carregar dados
df = pd.read_csv("../data/livros_vendidos.csv")

# Separar colunas categóricas e numéricas
cat_cols = ["genero", "autor", "editora"]
num_cols = ["preco"]

X = df[cat_cols + num_cols]
y = df["vendas"]

encoder = OrdinalEncoder()
X_xgb = X.copy()
X_xgb[cat_cols] = encoder.fit_transform(X[cat_cols])

model_xgb = XGBRegressor(n_estimators=300, learning_rate=0.1, max_depth=4, random_state=42)
model_xgb.fit(X_xgb, y)

# Novos livros para previsão
novos = pd.DataFrame(
    {
        "genero": ["Ciência"],
        "autor": ["Rafael Gomes"],
        "editora": ["Editora E"],
        "preco": [30],
    }
)

novos_xgb = novos.copy()
novos_xgb[cat_cols] = encoder.transform(novos[cat_cols])

pred_xgb = model_xgb.predict(novos_xgb)
print("XGBoost →", pred_xgb)
```
