```python
import pandas as pd
from catboost import CatBoostRegressor

# Carregar dados
df = pd.read_csv("../data/livros_vendidos.csv")

# Separar colunas categóricas e numéricas
cat_cols = ["genero", "autor", "editora"]
num_cols = ["preco"]

X = df[cat_cols + num_cols]
y = df["vendas"]

model_cat = CatBoostRegressor(iterations=500, learning_rate=0.05, depth=6, cat_features=cat_cols, verbose=False)

# Novos livros para previsão
novos = pd.DataFrame(
    {
        "genero": ["Ciência"],
        "autor": ["Rafael Gomes"],
        "editora": ["Editora E"],
        "preco": [30],
    }
)

model_cat.fit(X, y)

pred_cat = model_cat.predict(novos)
print("CatBoost →", pred_cat)
```
