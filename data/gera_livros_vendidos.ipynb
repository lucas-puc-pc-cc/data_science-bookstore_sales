```python
import pandas as pd
import random
import numpy as np

# fmt:off
# Mais temas, autores e editoras para aumentar a diversidade
temas = ["Ficção", "Romance", "Fantasia", "Terror", "Autoajuda", "Ciência", "História", "Tecnologia", "Biografia", "Infantil", "Aventura", "Suspense"]
autores = ["João Silva", "Maria Oliveira", "Carlos Souza", "Ana Pereira", "Lucas Lima", "Fernanda Costa", "Rafael Gomes", "Juliana Rocha", "Pedro Martins", "Carla Dias", "Sofia Almeida", "Thiago Fernandes", "Beatriz Nunes", "Gustavo Ribeiro"]
editoras = ["Companhia das Letras", "Editora Record", "Editora Planeta", "Editora Globo", "Editora Rocco", "Intrínseca", "Editora Sextante"]
livrarias = ["Livraria Cultura", "Livraria Saraiva", "Livraria da Travessa", "Livraria Martins Fontes", "Livraria Lecos", "Livraria Fnac"]

# Função para gerar nomes de livros
def gerar_nome_livro(genero, autor):
    if genero == "Romance": return f"Amor Eterno: {random.choice(['O Último Beijo', 'A Noite das Estrelas', 'Entre Nós', 'Corações em Chamas'])}"
    elif genero == "Ficção": return f"{random.choice(['A Última Fronteira', 'O Horizonte Proibido', 'Além do Infinito'])}"
    elif genero == "Fantasia": return f"O Feitiço da {random.choice(['Lua', 'Chama', 'Sombra', 'Estrela', 'Montanha'])}"
    elif genero == "Terror":return f"{random.choice(['A Casa das Sombras', 'No Coração do Terror', 'Os Olhos da Noite'])}"
    elif genero == "Autoajuda": return f"Desperte Seu Poder: {random.choice(['Superando Limites', 'Caminhos para a Felicidade', 'Vença o Medo'])}"
    elif genero == "Ciência": return f"Revolução Científica: {random.choice(['O Enigma da Matéria', 'Explorando o Infinito', 'Mistérios do Universo'])}"
    elif genero == "História": return f"{random.choice(['A Última Batalha', 'Nos Bastidores da História', 'Os Guerreiros do Passado'])}"
    elif genero == "Tecnologia": return f"Futuro Agora: {random.choice(['Revolução Digital', 'Inteligência Artificial', 'Conectando o Mundo'])}"
    elif genero == "Biografia": return f"Vida Incrível: {random.choice(['O Legado de uma Lenda', 'Histórias de Superação', 'Passos para a Eternidade'])}"
    elif genero == "Infantil": return f"Aventuras Mágicas: {random.choice(['O Segredo da Floresta Encantada', 'O Reino do Arco-Íris', 'A Fada e o Dragão'])}"
    elif genero == "Aventura": return f"Aventura Radical: {random.choice(['Em Busca do Tesouro Perdido', 'Os Mistérios do Deserto', 'Viagem ao Centro da Terra'])}"
    elif genero == "Suspense": return f"O Mistério de {random.choice(['Londres', 'Nova York', 'Paris', 'Moscou'])}"

# Gerando preços com mais variação usando log-normal
precos = np.random.lognormal(mean=3.5, sigma=0.8, size=1000)
precos = [round(min(p, 500), 2) for p in precos]  # limitando preço máximo a 500

# Parâmetros da distribuição Gamma (forma e escala)
# Esses valores podem ser ajustados para obter mais dispersão
shape = 2.0  # Forma
scale = 1000  # Escala

# Modificando a geração de vendas para aumentar a diferença entre autores

# Parâmetros da distribuição Gamma (forma e escala)
shape = 0.5  # Forma mais baixa para maior dispersão
scale = 2000  # Aumento da escala para gerar valores maiores

# Gerando as vendas com a nova distribuição Gamma
vendas = np.random.gamma(shape, scale, 1000)
# Clipping para limitar os valores extremos (e não deixar que se tornem absurdamente grandes)
vendas = np.clip(vendas, 15, np.percentile(vendas, 99))
# Normalizando para uma faixa de 0 a 250
vendas = (250 * (vendas - min(vendas)) / (max(vendas) - min(vendas))).astype(int)

# Aplicando uma variação maior por autor (aumentando os fatores de variação)
fatores_variacao = np.random.uniform(0.5, 5.0, 1000)  # Fatores de 0.5 a 5.0

# Aplicando a variação
vendas = (vendas * fatores_variacao).astype(int)

# Agora, crie o DataFrame novamente com as vendas ajustadas
df = pd.DataFrame({
    "genero": [random.choice(temas) for _ in range(1000)],
    "autor": [random.choice(autores) for _ in range(1000)],
    "preco": precos,
    "editora": [random.choice(editoras) for _ in range(1000)],
    "vendas": vendas,
})

# Gerando os nomes dos livros
df["livro"] = [gerar_nome_livro(row["genero"], row["autor"]) for _, row in df.iterrows()]
# Adicionando a coluna de livraria
df["livraria"] = [random.choice(livrarias) for _ in range(1000)]

# Selecionando as colunas finais
df = df[["livro", "genero", "autor", "editora", "preco", "livraria", "vendas"]]

# Salvando em CSV novamente
df.to_csv("livros_vendidos.csv", index=False)
```
