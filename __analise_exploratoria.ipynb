```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações visuais
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Supondo que seu DataFrame se chame df
df = pd.read_csv("../data/livros_vendidos2.csv")
print(df.columns)
df
```

# Analise exploratoria

```python
# 1. Inspeção inicial
print("----- 5 primeiras linhas -----")
print(df.head())
print("\n----- Informações do DataFrame -----")
print(df.info())
print("\n----- Estatísticas descritivas -----")
print(df.describe())
```

```python
# 2. Verificando valores ausentes
print("\n----- Valores ausentes por coluna -----")
print(df.isnull().sum())
```

```python
# 3. Distribuição de vendas e preços
plt.figure()
sns.histplot(df["NR_PRECO_VENDA"], bins=20, kde=True)
plt.title("Distribuição de Preços")
plt.xlabel("Preço")
plt.ylabel("Quantidade de livros")
plt.show()

plt.figure()
sns.histplot(df["NR_UNIDADES_VENDIDAS"], bins=20, kde=True)
plt.title("Distribuição de Vendas")
plt.xlabel("NR_UNIDADES_VENDIDAS")
plt.ylabel("Quantidade de livros")
plt.show()
```

```python
# 4. Top 10 autores por vendas
top_autores = df.groupby("NM_AUTOR")["NR_UNIDADES_VENDIDAS"].sum().sort_values(ascending=False).head(10)
plt.figure()
top_autores.plot(kind="bar", color="skyblue")
plt.title("Top 10 Autores por Vendas")
plt.ylabel("Vendas Totais")
plt.xlabel("NM_AUTOR")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Top 10 editoras por vendas
top_editoras = df.groupby("NM_EDITORA")["NR_UNIDADES_VENDIDAS"].sum().sort_values(ascending=False).head(10)
plt.figure()
top_editoras.plot(kind="bar", color="salmon")
plt.title("Top 10 Editoras por Vendas")
plt.ylabel("Vendas Totais")
plt.xlabel("NM_EDITORA")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Vendas médias por genero
vendas_por_genero = df.groupby("NM_GENERO")["NR_UNIDADES_VENDIDAS"].mean().sort_values(ascending=False)
plt.figure()
vendas_por_genero.plot(kind="bar", color="green")
plt.title("Vendas Médias por Tema")
plt.ylabel("Vendas Médias")
plt.xlabel("Tema")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Relação entre preço e vendas
plt.figure()
sns.scatterplot(data=df, x="NR_PRECO_VENDA", y="NR_UNIDADES_VENDIDAS", hue="NM_GENERO", alpha=0.7)
plt.title("Relação entre Preço e Vendas")
plt.xlabel("Preço")
plt.ylabel("NR_UNIDADES_VENDIDAS")
plt.show()
```

```python
# 8. Correlação numérica
plt.figure()
sns.heatmap(df[["NR_PRECO_VENDA", "NR_UNIDADES_VENDIDAS"]].corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de Correlação")
plt.show()
```

# VENDO SE TEM LINEARIDADE

```python
import matplotlib.pyplot as plt

plt.scatter(df["NR_PRECO_VENDA"], df["NR_UNIDADES_VENDIDAS"])
plt.xlabel("Preço")
plt.ylabel("NR_UNIDADES_VENDIDAS")
plt.title("Preço vs Vendas")
plt.show()

# Se os pontos formarem aproximadamente uma linha, há linearidade.
# Se houver curvas, platôs ou clusters, a relação é não-linear.
```

```python
import seaborn as sns

sns.boxplot(x="NM_GENERO", y="NR_UNIDADES_VENDIDAS", data=df)
plt.show()

sns.boxplot(x="NM_AUTOR", y="NR_UNIDADES_VENDIDAS", data=df)
plt.show()

sns.boxplot(x="NM_EDITORA", y="NR_UNIDADES_VENDIDAS", data=df)
plt.show()

# Esses gráficos mostram como categorias diferentes impactam vendas.
# Relações categóricas geralmente não são lineares, então modelos lineares podem não capturar isso bem.
```

```python
corr = df["NR_PRECO_VENDA"].corr(df["NR_UNIDADES_VENDIDAS"])
print("Correlação entre preço e vendas:", corr)

# Correlação próxima de ±1 sugere forte relação linear.
# Correlação perto de 0 sugere pouca ou nenhuma linearidade.
# Note: isso só funciona para variáveis numéricas.
```

```python
# 3. Transformar variáveis categóricas
# Se quiser testar linearidade considerando todas as features:
# Converta genero, autor e editora em one-hot encoding:
df_encoded = pd.get_dummies(df, columns=["livro", "NM_GENERO", "NM_AUTOR", "NM_EDITORA", "livraria"], drop_first=True)

# Depois, ajuste um modelo de regressão linear e veja os resíduos:
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df_encoded.drop("NR_UNIDADES_VENDIDAS", axis=1)
y = df_encoded["NR_UNIDADES_VENDIDAS"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

residuals = y - y_pred

plt.scatter(y_pred, residuals)
plt.xlabel("Predições")
plt.ylabel("Resíduos")
plt.title("Resíduos vs Predições")
plt.axhline(0, color="red")
plt.show()

# Se os resíduos se distribuírem aleatoriamente em torno de 0 → linearidade aceitável.
# Se houver curvas ou padrões → modelo linear não é adequado.
```
